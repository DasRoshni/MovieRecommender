# PROJECT 1:

Features taken into consideration while recommending list of movies is in the same order as below:
1.	Movies watched by users who have watched one or more movies same as input user
2.	Genres-If there exists other users who have watched a movie of the same genre as the input user, then suggest the input user with other genres watched by these users
3.	Rating – Suggest top 5 rated movies after filtering based on above 2 points. 

RDDS:
1. For list of movies and corresponding MovieIds
2. For list of MovieIds and corresponding genres of the movie
3. For list of movies and corresponding user ratings
4. To combine the ratings to find the resultant average per movie
5. To find the users for each movie

Flow:
1. Read data from movies.csv and ratings.csv file
2. Create dataset of userId and corresponding movie watched and movie rating given
3. Create dataset of MovieIds and corresponding genres of the movie
4. Create list of MovieIds and corresponding genres of the movie
5. Map each movie to the average ratings by all users
6. For the user we want to recommend movie to, take UserId and MovieId as input 
7. Filter down datasets according to movies and corresponsing genres watched by the user
8. Now find the users who have watched the same movie as input user and check what other movies these users have watched and their corresponding genres
9. Filter the dataset according to point 8 and order the list by descending average user rating.
10. Select top 5 from the resultant dataset
